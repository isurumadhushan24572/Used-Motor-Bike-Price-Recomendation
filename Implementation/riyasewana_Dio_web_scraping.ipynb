```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_riyasewana(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        bike_details = []
        listings = soup.find_all('li', class_='item round')

        for item in listings:
            title_link = item.find('h2', class_='more').find('a')
            img = item.find('img')
            location = item.find('div', class_='boxintxt').text.strip()
            price = item.find('div', class_='boxintxt b').text.strip()

            # Correctly extract the mileage information
            mileage_text = item.find_all('div', class_='boxintxt')[2].text.strip()
            mileage = ''.join(c for c in mileage_text if c.isdigit())

            date = item.find('div', class_='boxintxt s').text.strip()

            bike_details.append({
                'Title': title_link.get('title'),
                'Link': title_link.get('href'),
                'Location': location,
                'Price': price,
                'Mileage': mileage,
                'Date': date
            })

        return bike_details

    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return []

def scrape_all_pages(base_url, total_pages):
    all_bike_details = []

    for page_number in range(1, total_pages + 1):
        page_url = f"{base_url}&page={page_number}"
        print(f"Scraping page {page_number}...")

        page_bike_details = scrape_riyasewana(page_url)
        all_bike_details.extend(page_bike_details)

    df = pd.DataFrame(all_bike_details)
    df.to_excel('riyasewana_all_dio_bikes.xlsx', index=False)

    print("Scraping and saving to Excel completed.")

# URL of the webpage to scrape (without page parameter)
base_url_to_scrape = "https://riyasewana.com/search/honda/dio?"

# Set the total number of pages to scrape
total_pages_to_scrape = 49  # Adjust as needed

# Call the function with the base URL and total pages
scrape_all_pages(base_url_to_scrape, total_pages_to_scrape)
```
